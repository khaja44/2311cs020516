```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset("tips")
df
```

```python

df.describe
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python

df['smoker']=lb.fit_transform(df['smoker'])
df['sex']=lb.fit_transform(df['sex'])
df['time']=lb.fit_transform(df['time'])
df['day']=lb.fit_transform(df['day'])

```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
sns.heatmap(np.abs(df.corr()),cmap='Purples',annot=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
sns.heatmap(np.abs(df.corr())>0.17,cmap='Greens',annot = True)
plt.show()
```

Scaling techniques
1) standardization ----> mean 0 and std of 1
2) normalization  -----> range of 0 to 1 inclusive

```python
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
```

```python
scale_array=std_scaler.fit_transform(df)
```

```python
scale_array.shape
```

```python
scale_array
```

```python
scale_df=pd.DataFrame(scale_array,columns=df.columns)
scale_df.head(3)
```

```python
mx_array=mx_scaler.fit_transform(df)
```
