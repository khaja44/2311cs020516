# preprocessing

```python
df=pd.get_dummies(titanic,dtype='int')
df.head()
```

```python
titanic.shape[0]
```

# apriori algo

```python
freq = apriori(df, min_support=0.5, use_colnames=True)
freq
```

```python
rules = association_rules(freq,metric = 'lift',min_threshold=1,num_itemsets=len(freq))
rules
```

```python
rules.sort_values('lift',ascending=False)[0:5]
```
